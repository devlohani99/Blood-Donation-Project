```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

# Load the California Housing dataset
housing_data = fetch_california_housing(as_frame=True)

# Create DataFrame
df = housing_data.data
df['MedHouseVal'] = housing_data.target

# Compute correlation matrix
correlation_matrix = df.corr()

print(correlation_matrix)

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    square=True
)

plt.title("Correlation Matrix Heatmap")
plt.show()

# Pair Plot (using sample for faster execution)
sample_df = df.sample(1000, random_state=42)

sns.pairplot(
    sample_df,
    diag_kind='kde',
    corner=True
)

plt.show()
```

```
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

iris = load_iris()

X = StandardScaler().fit_transform(iris.data)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:,0], X_pca[:,1], c=iris.target)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Iris Dataset")
plt.show()
```

```
import pandas as pd

data = pd.read_csv("Find_S_Dataset.csv")

X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values

h = None

for i in range(len(Y)):
    if Y[i] == "Yes":
        if h is None:
            h = list(X[i])
        else:
            for j in range(len(h)):
                if h[j] != X[i][j]:
                    h[j] = '?'

print("Final Hypothesis:", h)
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

x = np.random.rand(100, 1)

train_x = x[:50]
train_y = [1 if i <= 0.5 else 2 for i in train_x.ravel()]

test_x = x[50:]
test_y = [1 if i <= 0.5 else 2 for i in test_x.ravel()]

predictions = {}

for k in [1, 2, 3, 4, 5, 20, 30]:
    knn = KNeighborsClassifier(n_neighbors=k)

    knn.fit(train_x, train_y)

    pred = knn.predict(test_x)
    predictions[k] = pred

    acc = np.mean(pred == test_y)

    print("k =", k, "Accuracy =", round(acc, 2))

k = 30
pred = predictions[k]

plt.scatter(train_x, train_y, label="Training Data")
plt.scatter(test_x, pred, marker='x', label="Predicted Test Data")

plt.xlabel("X Values")
plt.ylabel("Class")
plt.title("KNN Classification")
plt.legend()
plt.show()
```

```
import numpy as np, matplotlib.pyplot as plt

def l(x,y,z,t):
    w=np.diag(np.exp(-(x-z)**2/(2*t*t)))
    a=np.c_[np.ones(len(x)),x]
    return [1,z]@(np.linalg.pinv(a.T@w@a)@a.T@w@y)

x=np.linspace(0,10,100)
y=np.sin(x)

plt.scatter(x,y)
plt.plot(x,[l(x,y,i,.5) for i in x],'r')
plt.show()
```

```
//7
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

df = pd.read_csv("auto_mpg.csv").dropna()

X = df[['horsepower']]
y = df['mpg']

model = make_pipeline(PolynomialFeatures(3), LinearRegression())
model.fit(X, y)

plt.scatter(X, y)
plt.plot(X, model.predict(X), 'r')
plt.show()
```

```
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

data = load_breast_cancer()

model = DecisionTreeClassifier(max_depth=3)
model.fit(data.data, data.target)

plt.figure(figsize=(12,8))
plot_tree(
    model,
    feature_names=data.feature_names,
    class_names=data.target_names,
    filled=True
)
plt.show()

sample = data.data[0].reshape(1,-1)
print("Prediction:", data.target_names[model.predict(sample)[0]])
```

```
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import *
import matplotlib.pyplot as plt

X,Y=fetch_olivetti_faces(return_X_y=True)

print("Dataset Shape:",X.shape)
print("Number of Classes:",40)

X1,X2,Y1,Y2=train_test_split(X,Y,test_size=0.2,random_state=42)
P=GaussianNB().fit(X1,Y1).predict(X2)

print("Accuracy:",round(accuracy_score(Y2,P),2))
print(classification_report(Y2,P))

f,a=plt.subplots(5,8)
for i,x in enumerate(a.flat):
    x.imshow(X[i*10].reshape(64,64),cmap="gray")
    x.axis("off")
plt.show()
```

```
//10. Develop a program to implement k-means clustering using
//Wisconsin Breast Cancer data set and visualize the clustering result.


from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X=load_breast_cancer().data
k=KMeans(2).fit(X)
p=PCA(2).fit(X)

plt.scatter(*p.transform(X).T,c=k.labels_)
plt.scatter(*p.transform(k.cluster_centers_).T,c='r',marker='X',s=200)
plt.show()
```
